## **Convertors**

```python
def rename(name):
  if name=="Gaddafi Stadium, Lahore":
    return "GSL"

  else:
    return name
```

```python
rename("Gaddafi Stadium, Lahore")
```

```python
pd.read_csv('PSL_Season_09_dataset.csv',converters={'venue':rename})
```
